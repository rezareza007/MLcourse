```python
import matplotlib.pyplot as plt
import numpy as np
```

## plot

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x),label="line1")
plt.plot(x, np.cos(x),label="line2")
plt.plot(x,np.sin(x)+np.cos(x),label="line3")
plt.title("My first figure") 
plt.xlabel("My x-axis")   
plt.ylabel("My y-axis")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(2,2)
print(ax.shape)
ax[0,0].plot(np.arange(6))   
ax[0,1].plot(np.arange(6,0,-1))    
ax[1,0].plot((-1)**np.arange(6))    
ax[1,1].plot((-1)**np.arange(1,7))
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

#size of the plot
plt.figure(figsize=(9, 3))

# says that 1*3 plot and the first one
plt.subplot(131)
plt.bar(names, values)
# says that 1*3 plot and the second one
plt.subplot(132)
plt.scatter(names, values)
# say that 1*3 plot and the third one
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

## scatter plot

## pie

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0.15, 0.1, 0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal') 
plt.show()
```

## conture

```python
x = np.arange(1, 10)
y = x.reshape(-1, 1)
h = x * y

cs = plt.contourf(h, levels=[10, 30, 50],
    colors=['#808080', '#A0A0A0', '#C0C0C0'], extend='both')
cs.cmap.set_over('red')
cs.cmap.set_under('blue')
cs.changed()
```

```python
delta = 0.25
x = np.arange(-3.0, 5.0, delta)
y = np.arange(-1.3, 2.5, delta)
X, Y = np.meshgrid(x, y)
Z = (np.exp(-X**2 - Y**2) - np.exp(-(X - 1)**2 - (Y - 1)**2))
   
fig, ax = plt.subplots()
im = ax.imshow(Z, interpolation ='bilinear',
               origin ='lower',
               cmap ="bone", 
               extent =(-3, 3, -2, 2))
  
levels = np.arange(-1.2, 1.6, 0.2)
CS = ax.contour(Z, levels, 
                origin ='lower',
                cmap ='Greens',
                linewidths = 2,
                extent =(-3, 3, -2, 2))
  
zc = CS.collections[6]
plt.setp(zc, linewidth = 2)
   
#  write number
ax.clabel(CS, levels,
          inline = 1, 
          fmt ='% 1.1f',
          fontsize = 14)
  
plt.title('matplotlib.pyplot.contour() Example')
plt.show()
```

## fill

```python
def f(x):
    return x**2

x = np.arange(0,10,0.1)
y = f(x)

plt.plot(x,y,'k--')

plt.fill_between(x, y, np.max(y), color='#539ecd')

plt.grid()

plt.title('How to fill an area in matplotlib ?',fontsize=10)
plt.show()
```
